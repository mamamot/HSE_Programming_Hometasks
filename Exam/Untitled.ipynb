```python
import pandas as pd
import numpy as np
```

```python
with open("source_post1950_wordcount.xls", "rb") as xls:
    data = pd.read_excel(xls)
```

```python
data = data[data['words']!='none']
```

```python
data['words'] = data['words'].apply(int)
```

```python
ids_to_drop = list()
new_entries = list()
for i, row in data.iterrows():
    if row['words'] > 70000 and row['words'] < 100000:
        new_1 = row.copy()
        new_2 = row.copy()
        new_1.words = row.words / 2
        new_2.words = row.words - new_1.words
        new_1.path = new_1.path + "_1"
        new_2.path = new_2.path + "_2"
        ids_to_drop.append(row.path)
        new_entries.append(new_1)
        new_entries.append(new_2)
    if row['words'] >= 100000:
        new_1 = row.copy()
        new_2 = row.copy()
        new_3 = row.copy()        
        new_1.words = row.words / 3
        new_2.words = row.words / 3
        new_3.words = row.words - new_1.words - new_2.words
        new_1.path = new_1.path + "_1"
        new_2.path = new_2.path + "_2"
        new_3.path = new_3.path + "_2"
        ids_to_drop.append(row.path)
        new_entries.append(new_1)
        new_entries.append(new_2)
        new_entries.append(new_3)
```

```python
ids_to_drop
```

```python
for id_to_drop in ids_to_drop:
    data = data[data['path']!=id_to_drop]
```

```python
for new_entry in new_entries:
    data = data.append(new_entry)
```

```python
needed_spheres = {
    'художественная': 40000000,
    'публицистика': 29000000,
    # "мемуары'
    'учебно-научная': 12000000,
    'официально-деловая': 1500000,
    'церковно-богословская': 1500000,
    'бытовая': 1800000,
    'электронная коммуникация': 1800000,
    'производственно-техническая': 900000,
    'реклама': 800000
}
```

```python
cut_down_by_spheres = pd.DataFrame(columns=['path', 'author', 'sex', 'birthday', 'header', 'created', 'sphere',
       'genre_fi', 'type', 'topic', 'chronotop', 'style', 'audience_age',
       'audience_level', 'audience_size', 'source', 'publication', 'publisher',
       'publ_year', 'medium', 'subcorpus', 'tagging', 'words', 'Unnamed: 23'])
```

```python
for sphere in needed_spheres:
    total = 0
    index = 0
    sphere_data = data[data['sphere']==sphere]
    while total < needed_spheres[sphere] and index < len(sphere_data):
        sphere_row = sphere_data.iloc[index]
        cut_down_by_spheres = cut_down_by_spheres.append(sphere_row)
        total += sphere_row['words']
```

```python
cut_down_by_spheres.to_csv("spheres_cut.csv", index=False)
```
